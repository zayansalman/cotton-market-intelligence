# Strategic procurement demo

Given **X tonnes** over **Y months**, builds a **month-by-month roadmap** using price signals, volatility, RSS headlines (`config/news_feeds.yml`), and an optional LLM narrative (`OPENAI_API_KEY`).

Place `data/cotton_macrotrends_daily.csv` or set paths below.

```python
from pathlib import Path
import json

from src.strategic import build_strategic_procurement_plan, plan_to_dict

project_root = Path("..").resolve()
macrotrends_csv = project_root / "data" / "cotton_macrotrends_daily.csv"
worldbank_xlsx = project_root / "data" / "WBCOMM_Prices.xlsx"

if not macrotrends_csv.exists():
    raise FileNotFoundError(f"Need MacroTrends CSV at {macrotrends_csv}")

wb = str(worldbank_xlsx) if worldbank_xlsx.exists() else None

plan = build_strategic_procurement_plan(
    company="Demo Mill",
    total_tonnes=5000.0,
    horizon_months=6,
    macrotrends_csv=macrotrends_csv,
    worldbank_xlsx=wb,
    mill_profile_name=None,
)

out = plan_to_dict(plan)
print(json.dumps(out, indent=2)[:4000])
```

```python
# Month-by-month tonnes
for row in out["roadmap"]:
    print(f"Month {row['month']}: {row['tonnes']:.1f} t — {row['note']}")
```
